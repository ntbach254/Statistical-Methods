```python
def find_outliers(column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    return outliers

# Finding outliers for Sleep Duration, Quality of Sleep, and Stress Level
outliers_sleep_duration = find_outliers('Sleep Duration')
outliers_quality_of_sleep = find_outliers('Quality of Sleep')
outliers_stress_level = find_outliers('Stress Level')

print("Outliers in Sleep Duration:\n", outliers_sleep_duration)
print("Outliers in Quality of Sleep:\n", outliers_quality_of_sleep)
print("Outliers in Stress Level:\n", outliers_stress_level)
```

```python
# Calculate the correlation coefficient
correlation_sleep_duration = df['Quality of Sleep'].corr(df['Sleep Duration'])
correlation_stress_level = df['Quality of Sleep'].corr(df['Stress Level'])

print(f'Correlation between Sleep Duration and Quality of Sleep: {correlation_sleep_duration}')
print(f'Correlation between Stress Level and Quality of Sleep: {correlation_stress_level}')
```

```python
from scipy import stats

# Least squares regression for Sleep Duration vs. Quality of Sleep
slope_sd, intercept_sd, r_value_sd, p_value_sd, std_err_sd = stats.linregress(df['Sleep Duration'], df['Quality of Sleep'])
print(f'Regression equation for Quality of Sleep: Quality of Sleep = {slope_sd:.2f} * Sleep Duration + {intercept_sd:.2f}')
print(f'R-squared for Sleep Duration: {r_value_sd**2:.2f}')

# Plot the regression line for Sleep Duration
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df['Sleep Duration'], y=df['Quality of Sleep'])
plt.plot(df['Sleep Duration'], slope_sd * df['Sleep Duration'] + intercept_sd, color='red')
plt.title('Sleep Duration vs. Quality of Sleep with Regression Line')
plt.xlabel('Sleep Duration')
plt.ylabel('Quality of Sleep')
plt.show()

# Least squares regression for Stress Level vs. Quality of Sleep
slope_sl, intercept_sl, r_value_sl, p_value_sl, std_err_sl = stats.linregress(df['Stress Level'], df['Quality of Sleep'])
print(f'Regression equation for Quality of Sleep: Quality of Sleep = {slope_sl:.2f} * Stress Level + {intercept_sl:.2f}')
print(f'R-squared for Stress Level: {r_value_sl**2:.2f}')

# Plot the regression line for Stress Level
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df['Stress Level'], y=df['Quality of Sleep'])
plt.plot(df['Stress Level'], slope_sl * df['Stress Level'] + intercept_sl, color='red')
plt.title('Stress Level vs. Quality of Sleep with Regression Line')
plt.xlabel('Stress Level')
plt.ylabel('Quality of Sleep')
plt.show()
```

```python
from scipy.stats import shapiro
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats

# Shapiro-Wilk test for Sleep Duration
stat, p = shapiro(df['Sleep Duration'])
print('Shapiro-Wilk Test for Sleep Duration: Statistics=%.3f, p=%.3f' % (stat, p))

# Q-Q plot for Sleep Duration
plt.figure(figsize=(10, 6))
stats.probplot(df['Sleep Duration'], dist="norm", plot=plt)
plt.title('Q-Q Plot for Sleep Duration')
plt.show()

# Shapiro-Wilk test for Stress Level
stat, p = shapiro(df['Stress Level'])
print('Shapiro-Wilk Test for Stress Level: Statistics=%.3f, p=%.3f' % (stat, p))

# Q-Q plot for Stress Level
plt.figure(figsize=(10, 6))
stats.probplot(df['Stress Level'], dist="norm", plot=plt)
plt.title('Q-Q Plot for Stress Level')
plt.show()
```

```python
import numpy as np

# Confidence interval for Sleep Duration
mean_sleep_duration = np.mean(df['Sleep Duration'])
std_sleep_duration = np.std(df['Sleep Duration'], ddof=1)
n_sleep_duration = len(df['Sleep Duration'])
confidence_level = 0.95
alpha = 1 - confidence_level
t_critical = stats.t.ppf(1 - alpha/2, df=n_sleep_duration-1)

margin_of_error_sleep = t_critical * (std_sleep_duration/np.sqrt(n_sleep_duration))
confidence_interval_sleep = (mean_sleep_duration - margin_of_error_sleep, mean_sleep_duration + margin_of_error_sleep)

print(f'95% Confidence Interval for Sleep Duration: {confidence_interval_sleep}')

# Confidence interval for Stress Level
mean_stress_level = np.mean(df['Stress Level'])
std_stress_level = np.std(df['Stress Level'], ddof=1)
n_stress_level = len(df['Stress Level'])

margin_of_error_stress = t_critical * (std_stress_level/np.sqrt(n_stress_level))
confidence_interval_stress = (mean_stress_level - margin_of_error_stress, mean_stress_level + margin_of_error_stress)

print(f'95% Confidence Interval for Stress Level: {confidence_interval_stress}')
```

```python
import pandas as pd
import statsmodels.api as sm
from scipy import stats

# Load the dataset
df = pd.read_csv('https://raw.githubusercontent.com/ntbach254/Statistical-Methods/refs/heads/main/dataset/Sleep_health_and_lifestyle_dataset%20-%20Sleep_health_and_lifestyle_dataset.csv')

# Regression Model 1: Sleep Duration vs. Quality of Sleep
X_sd = df['Sleep Duration']
y_qs = df['Quality of Sleep']
X_sd = sm.add_constant(X_sd)  # Adds a constant term to the predictor

# Fit the regression model
model_sd = sm.OLS(y_qs, X_sd).fit()

# Extract F-statistic and degrees of freedom
f_statistic_sd = model_sd.fvalue
dfn_sd = model_sd.df_model
dfd_sd = model_sd.df_resid

# Calculate critical F-value
alpha = 0.05
f_critical_sd = stats.f.ppf(1 - alpha, dfn_sd, dfd_sd)

# Calculate p-value
f_p_value_sd = model_sd.f_pvalue

print(f"F-statistic (Sleep Duration): {f_statistic_sd:.2f}")
print(f"Critical F-value (Sleep Duration): {f_critical_sd:.2f}")
print(f"P-value (Sleep Duration): {f_p_value_sd:.4f}")

# Decision rule for Sleep Duration model
if f_statistic_sd > f_critical_sd:
    print("Reject null hypothesis for Sleep Duration, there is a significant regression relation.")
else:
    print("Fail to reject null hypothesis for Sleep Duration, there is no significant regression relation.")

```

```python
# Regression Model 2: Stress Level vs. Quality of Sleep
X_sl = df['Stress Level']
y_qs = df['Quality of Sleep']
X_sl = sm.add_constant(X_sl)  # Adds a constant term to the predictor

# Fit the regression model
model_sl = sm.OLS(y_qs, X_sl).fit()

# Extract F-statistic and degrees of freedom
f_statistic_sl = model_sl.fvalue
dfn_sl = model_sl.df_model
dfd_sl = model_sl.df_resid

# Calculate critical F-value
f_critical_sl = stats.f.ppf(1 - alpha, dfn_sl, dfd_sl)

# Calculate p-value
f_p_value_sl = model_sl.f_pvalue

print(f"F-statistic (Stress Level): {f_statistic_sl:.2f}")
print(f"Critical F-value (Stress Level): {f_critical_sl:.2f}")
print(f"P-value (Stress Level): {f_p_value_sl:.4f}")

# Decision rule for Stress Level model
if f_statistic_sl > f_critical_sl:
    print("Reject null hypothesis for Stress Level, there is a significant regression relation.")
else:
    print("Fail to reject null hypothesis for Stress Level, there is no significant regression relation.")

```

```python
# Hypothesis test for Sleep Duration (H0: mean = hypothesized value)
hypothesized_mean_sleep = 7  # for example
t_statistic_sleep, p_value_sleep = stats.ttest_1samp(df['Sleep Duration'], hypothesized_mean_sleep)

print(f'Test Statistics for Sleep Duration: {t_statistic_sleep}')
print(f'P-value for Sleep Duration: {p_value_sleep}')

# Hypothesis test for Stress Level (H0: mean = hypothesized value)
hypothesized_mean_stress = 5  # for example
t_statistic_stress, p_value_stress = stats.ttest_1samp(df['Stress Level'], hypothesized_mean_stress)

print(f'Test Statistics for Stress Level: {t_statistic_stress}')
print(f'P-value for Stress Level: {p_value_stress}')
```

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('https://raw.githubusercontent.com/ntbach254/Statistical-Methods/refs/heads/main/dataset/Sleep_health_and_lifestyle_dataset%20-%20Sleep_health_and_lifestyle_dataset.csv')

# Define the explanatory and response variables
X = df['Stress Level']
y = df['Quality of Sleep']

# Add a constant term to the explanatory variable for the intercept
X = sm.add_constant(X)

# Fit the least squares regression model
model = sm.OLS(y, X).fit()

# Print the model summary
print(model.summary())

# Extract regression coefficients
intercept = model.params[0]
slope = model.params[1]

print(f"Intercept: {intercept}")
print(f"Slope: {slope}")

# Plot the regression line
plt.scatter(df['Quality of Sleep'], df['Stress Level'], color='blue', label='Data points')
plt.plot(df['Quality of Sleep'], intercept + slope * df['Quality of Sleep'], color='red', label='Regression line')
plt.xlabel('Quality of Sleep')
plt.ylabel('Stress Level')
plt.title('Regression Line: Stress Level vs Quality of Sleep')
plt.legend()
plt.show()
```
